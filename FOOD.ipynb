```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
files.upload()
```

```python
food=pd.read_csv('Foodpanda Analysis Dataset.csv')
food
```

```python
food.isnull().sum()
```

```python
food["churned"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in food.columns:
  if food[col].dtype=='object':
     food[col]=le.fit_transform(food[col])
```

```python
food
```

```python
print(food.head())
print(food.info())
```

```python
date_cols = ['signup_date', 'order_date', 'last_order_date', 'rating_date']
for col in date_cols:
    food[col] = pd.to_datetime(food[col], errors='coerce')
```

```python
sns.countplot(x='gender', data=food, palette='Set2')
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='age', data=food, palette='Set3')
plt.title("Age Group Distribution")
plt.show()
```

```python
categorical_cols = ['gender', 'age', 'city', 'restaurant_name', 'dish_name', 'category', 'payment_method', 'delivery_status', 'churned', 'rating']
for col in categorical_cols:
    print(f"\n Value counts for {col}:")
    print(food[col].value_counts())
```

```python
sns.countplot(data=food, x='gender')
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(data=food, x='churned')
plt.title("Churned vs Active Customers")
plt.show()
```

```python
sns.countplot(data=food,x='payment_method',order=food['payment_method'].value_counts().index)
plt.title("Payment Method Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=food, x='delivery_status')
plt.title("Delivery Status Distribution")
plt.show()
```

```python
sns.histplot(food['rating'], bins=5, kde=True)
plt.title("Customer Rating Distribution")
plt.show()
```

```python
sns.boxplot(data=food,x='churned', y='loyalty_points')
plt.title("Loyalty Points by Churn Status")
plt.show()
```

```python
sns.histplot(food['order_frequency'],bins=30, kde=True)
plt.title("Order Frequency Distribution")
plt.show()
```

```python
numerical_cols = food.select_dtypes(include=np.number).columns
correlation_matrix = food[numerical_cols].corr()
plt.figure(figsize=(22,10))
sns.heatmap(correlation_matrix,
            annot=True,         # Show correlation coefficients
            fmt='.2f',          # Format decimals
            cmap='coolwarm',    # Color scheme
            square=True,        # Keep squares
            linewidths=0.5,     # Lines between squares
            cbar_kws={"shrink": .8})  # Color bar size

plt.title('Correlation Heatmap of Numerical Features', fontsize=22)
plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['gender', 'age', 'city', 'restaurant_name', 'dish_name', 'category', 'payment_method', 'delivery_status', 'churned', 'rating']
for col in categorical_cols:
    plt.figure(figsize=(12, 6))
    order=food[col].value_counts().index
    sns.countplot(data=food, x=col, order=order, palette="Set2", hue=col, legend=False)
    plt.title(f'Count Plot of {col.capitalize()}', fontsize=14)
    plt.xlabel(col.capitalize())
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
print("City Count:\n",food['city'].value_counts(),"\n")
print("Category Count:\n",food['category'].value_counts(),"\n")
print("Dish Name Count:\n",food['dish_name'].value_counts(),"\n")
print("Rating Count:\n",food['rating'].value_counts().sort_index(),"\n")
```

```python
plt.figure(figsize=(12, 6))
top_dishes=food['dish_name'].value_counts().head(10).index
sns.countplot(data=food[food['dish_name'].isin(top_dishes)], x='dish_name', hue='churned', palette='Set1')
plt.title("Top 10 Dishes by Churn Status")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
